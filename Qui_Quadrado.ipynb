O Qui quadrado é utilizado para saber se existe alguma diferença significativa entre os dados analisado agora ou se essa diferença é fruto do acaso.

```python
import numpy as np
from scipy.stats import chi2_contingency
```

```python
#Criação da matriz com os dados
#19 homens que assistem, 6 que não assistem ,46 mulheres que assistem contra 32 mulheres que não assistem.
novela = np.array([[19 , 6] , [46 , 32]])
novela
```

```python
# Segundo valor é o pvalue
# Valor de p > 0,05 não há evidências de diferenças significativas ( hipótese nula) : não há diferença significativa
#entre homens e mulheres que assistem novela
chi2_contingency(novela)
```

```python
novela2 = np.array([[22,3] , [43 , 32]])
novela2
```

```python
#Valor de p < 0,05 , rejeita-se a hipótese nula e parte para a hipótese alternativa : há diferença significativa.
chi2_contingency(novela2)
```
