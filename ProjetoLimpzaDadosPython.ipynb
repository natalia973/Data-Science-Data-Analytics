```python
import pandas as pd
import statistics as sts
import numpy as np
import seaborn as sbn
```

```python
from os import read
tabela = pd.read_csv('/content/tempo.csv' , sep=";")
tabela
#A aparência deve ser sol, nublado, chuva.
#Temperatura de -130 até 130
#Umidade de 0 até 100
#Jogar - sim/não
#Tratar os dados NA - Non avalilable
```

```python
#Aqui, verifica-se a moda, os valores mais frequentes para então encaixar 'menos', que está fora do padrão requerido
agrupado = tabela.groupby(['Aparencia']).size()
agrupado
```

```python
#O valor 'menos' irá para o conjunto 'sol'
tabela.loc[tabela['Aparencia'] == 'menos' , 'Aparencia'] = 'sol'
agrupar = tabela.groupby(['Aparencia']).size()
agrupar
```

```python
#Filtrar a coluna temperatura para valores de -130 até 130
filtro= tabela.loc[(tabela['Temperatura'] >  -130) & (tabela['Temperatura'] <= 130)]

filtro
```

```python
#Filtrar umidade de 0 até 100:
filtro = tabela.loc[(tabela['Umidade'] > 0) & (tabela['Umidade'] <= 100)]
filtro
```

```python
agrupar = tabela.groupby(['Jogar']).size()
agrupar
```
