```python
#MANIPULANDO DADOS:
df = pd.read_csv('/content/coffee.csv')
df
```

```python
#Estabelecer um preço, comum a todos.
df['price'] = 5.00
df
```

```python
#Adicionar o preço do café com base no tipo.Expresso custa 5.0 e Latte custa 7.0
def adicionar_preco_cafe(nome):
  if nome == 'Espresso':
    return 5.0
  else:
      return 7.0
```

```python
df['price'] = df['Coffee Type'].apply(adicionar_preco_cafe)
df
```

```python
#Fazer a coluna de todas as receitas:
df['revenue'] = df['Units Sold'] * df['price']
df
```

```python
df.drop(columns=["revenue"],inplace=True)
```

```python
# df = df.drop(columns=['revenue'])
# df
```

```python
df['revenue'] = df['Units Sold'] * df['price']
df.loc[0:6,'Units Sold'] = np.nan
df.loc[0:6 , 'revenue'] = np.nan

#Substituindo as linhas de index 0 até o 6 por 'Nan'
```

```python
df
```

```python
df['Units Sold'] = df['Units Sold'].fillna(0)
df
```

```python
df['revenue'] = df['revenue'].fillna(df['revenue'].mean())
df
```

```python
#Agrupamento:
df.groupby(['Coffee Type'])['Units Sold'].sum()
#Agrupamento dos tipos de café pelo total de vendas de cada tipo.
```

```python
df.groupby(['Day'])['Units Sold'].sum()
```

```python
df.groupby(['Coffee Type'])['Units Sold'].mean()
```

```python
df.groupby(['Coffee Type']).agg({'Units Sold' : 'sum' , 'revenue' : 'mean'})
#Agrupamento com agregação da coluna cofee type com a variável soma das unidades vendidas mais a média das receitas de cada tipo.
```
