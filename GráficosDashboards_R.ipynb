```r
#Agregação de tabelas diferentes:
spray = aggregate(. ~ spray , data = InsectSprays , sum)
spray
```

```r
#Gráfico de barras:
barplot(spray$count, col = gray.colors(6), xlab = "Sprays" , ylab="Total" , names.arg = spray$sprayspray)
```

```r
#Gráfico de setor
pie(spray$count, labels = NA, main="Spray", col=c(1:6))
legend("bottomright", legend=spray$spray , cex = 1, fill=c(1:6))
```
