```python
import pandas as pd
import seaborn as srn
import numpy as np
import matplotlib.pyplot as plt
import statistics as sts
```

```python
df = pd.read_csv("/content/dados.csv", sep=';')
```

```python
df.info()
```

```python
df.head()
```

```python
#Qual a mdia do PIB de todas as cidades?
media=df['PIB'].mean()
#A mdia do PIB  19808.74133... Percebe-se que a casa decimal  muito grande, ser necessrio aproxim-la
media
```

```python
#Aproximao da mdia:
media = df['PIB'].mean().round(2)
media
```

```python
#Anlise de qual municpio tem maior desvio padro  no PIB:
DesvioP = df['PIB'].std()
DesvioP
```

```python
df.groupby('MUNICIPIO')['PIB'].std().argmax()
#Aqui retorna o ID do municpio com maior desvio padro
```

```python
#Nome do municpio com maior desvio padro:
#Primeiro foi agrupado os municipios pelo PIB, com o desvio P. do PIB.
serie_maior_dp = df.groupby('MUNICIPIO')['PIB'].std()
#Depois foi calculado com o argmax() o  maior valor:
Municipio_maior_dp = serie_maior_dp.argmax()

#Depois foi coleatado o ndice, para ser acessado o nome.
nomeMaior = serie_maior_dp.index[Municipio_maior_dp]
print(f'O municpio com maior desvio padro  o {nomeMaior}')
```

```python
#Anlise da coluna VALOREMPENHO : Mdia da coluna, aproximada duas casas
mediaVE = df['VALOREMPENHO'].mean().round(2)
mediaVE
```

```python
maiorVE = df.groupby('MUNICIPIO')['VALOREMPENHO'].mean()
maiorVE_index = maiorVE.argmax()
muniMAIOR = maiorVE.index[maiorVE_index]
print(f'O municpio com o maior VALOR EMPENHO  o {muniMAIOR}')
```
