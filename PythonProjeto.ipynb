```python
import pandas as pd
import seaborn as srn
import numpy as np
import matplotlib.pyplot as plt
import statistics as sts
```

```python
df = pd.read_csv("/content/dados.csv", sep=';')
```

```python
df.info()
```

```python
df.head()
```

```python
#Qual a média do PIB de todas as cidades?
media=df['PIB'].mean()
#A média do PIB é 19808.74133... Percebe-se que a casa decimal é muito grande, será necessário aproximá-la
media
```

```python
#Aproximação da média:
media = df['PIB'].mean().round(2)
media
```

```python
#Análise de qual município tem maior desvio padrão  no PIB:
DesvioP = df['PIB'].std()
DesvioP
```

```python
df.groupby('MUNICIPIO')['PIB'].std().argmax()
#Aqui retorna o ID do município com maior desvio padrão
```

```python
#Nome do município com maior desvio padrão:
#Primeiro foi agrupado os municipios pelo PIB, com o desvio P. do PIB.
serie_maior_dp = df.groupby('MUNICIPIO')['PIB'].std()
#Depois foi calculado com o argmax() o  maior valor:
Municipio_maior_dp = serie_maior_dp.argmax()

#Depois foi coleatado o índice, para ser acessado o nome.
nomeMaior = serie_maior_dp.index[Municipio_maior_dp]
print(f'O município com maior desvio padrão é o {nomeMaior}')
```

```python
#Análise da coluna VALOREMPENHO : Média da coluna, aproximada duas casas
mediaVE = df['VALOREMPENHO'].mean().round(2)
mediaVE
```

```python
maiorVE = df.groupby('MUNICIPIO')['VALOREMPENHO'].mean()
maiorVE_index = maiorVE.argmax()
muniMAIOR = maiorVE.index[maiorVE_index]
print(f'O município com o maior VALOR EMPENHO é o {muniMAIOR}')
```

```python
#Fazer um histplot do PIB dos 10 primeiros municípios:
agrupado = df.loc[:, ['PIB', 'MUNICIPIO']].head(10)
agrupado.plot.bar(x = 'MUNICIPIO' , y = 'PIB')

#Já encontramos os primeiros 10 valores do PIB e do município.
```
