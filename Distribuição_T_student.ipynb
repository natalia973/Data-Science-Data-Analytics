```python
from scipy.stats import t
```

Distribuição T student é usada primordialmente para amostras pequenas.

Suponha que a média dos salários dos cientistas de dados é R$ 75,00 por hora, de uma amostra com 9 funcionários e com desvio padrão de 10.

Qual a probabilidade de selecionar um cientista de dados e o salário ser menor que 80 por hora?  t = X -Mo / (Desvio padrão da amostra / raíz da amostra)

t = 1,5

```python

t.cdf(1.5 , 8)
#cdf - esquerda da distribuição , graus de liberdade = 8 = tamanho da amostra - 1
```

```python
t.sf(1.5 , 8)
#sf - a direita da distribuição
```

```python
#Somatório da distribuição da direita + esquerda = 1
t.cdf(1.5 , 8) + t.sf(1.5 , 8)
```
