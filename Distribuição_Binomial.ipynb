Para ser Distribuição Binomial :

*   Número fixo de experimentos.
*   Cada experimento pode ter ou sucesso ou fracasso.
*   A probabilidade de sucesso deve ser a mesma em cada elemento.
*   Experimentos são independentes.








Se você fizer a prova de um concurso com 12 questões. Chutando todas as questões, qual a probabilidade de acertar 7 questões? 4 alternativas cada questão?
X = 7 acertos
n = 12
p = 0,25 (1/4)
* 7 acertos e 5 fracassos  , p = 0,25 (0,75)

```python
from scipy.stats import binom
```

```python
#Se jogar uma moeda 5 vezes, qual a probabilidade de dar cara 3 vezes?
#Eventos , experimentos, probabilidades - Eventos - quantidade desejada , experimentos - total ,
#probabilidades - probabilidade de acontecer o evento
prob = binom.pmf(3 , 5 , 0.5)
prob
```

```python
#Passar por 4 sinais de 4 tempos, qual a probabilidade de pegar sinal verde
#nenhuma vez, 1 vez , 2 vezes, 3 vezes, 4 vezes?
prob = binom.pmf(0, 4, 0.25) + binom.pmf(1 , 4, 0.25) + binom.pmf(2 , 4 , 0.25 ) + binom.pmf(3 , 4 , 0.25) + binom.pmf(4 , 4 , 0.25)
prob
```

```python
# E se forem sinais de dois tempos?
prob = binom.pmf(4 , 4 , 0.5)
prob
```

```python
# Probabilidade acumulativa
prob = binom.cdf(4 , 4 , 0.5)
prob
```

```python
# Concurso com 12 questões, qual a probabilidade de acertar 7 questões sendo que cada questão possui 4 alternativas?
prob = binom.pmf(7, 12 , 0.25) * 100
prob
```

```python
# Probabilidade de acertar as 12 questões?
prob = binom.pmf(12 , 12, 0.25)
prob
```
