Com a distribuição de Poisson há a possibilidade de a probabilidade ser igual a média P(X = x).Captura de tela 2025-11-05 213222.png
Em que x = número de eventos que estão sendo calculados, lambda = número médio de eventos que ocorre por intervalo , e = constante de euler

A distribuição de Poisson mede a probabilidade da ocorrência de eventos em um intervalo de tempo, em vez de um certo número de experimentos, os eventos a cada intervalo devem ser independentes.

```python
from scipy.stats import poisson
#Média de acidentes de carro é 2 por dia.
```

```python
#Qual a probabilidade de ocorrerem 3 acidentes por dia?
poisson.pmf(3,2)
```

```python
#Qual a probabilidade de ocorrerem 3 ou menos acidentes por dia?
poisson.cdf(3,2)
```

```python
#Qual a probabilidade de ocorrerem mais de 3 acidentes por dia?
poisson.sf(3,2)
```
